```python
# from http://docs.opencv.org/master/db/d5c/tutorial_py_bg_subtraction.html#gsc.tab=0

import numpy as np
import cv2

cap = cv2.VideoCapture('/Users/pflomacpro/RLG/Data/15-09-03/Level18/Trim001.m4v')

fgbg = cv2.BackgroundSubtractorMOG(history=10, nmixtures=1, backgroundRatio=0.001,noiseSigma=1500.0)


while(1):
    ret, frame = cap.read()

    fgmask = fgbg.apply(frame)

    cv2.imshow('frame',fgmask)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
